**Slicing**

You can extract a portion of a list using the [start:stop:step]

syntax:

```python
numbers = [0, 1, 2, 3, 4, 5]
print(numbers[1:4])  # Output: [1, 2, 3] (index 1 to 3)
print(numbers[::-1])
```

```python
colors = ["red", "green", "blue", "yellow", "purple"]

print(colors[1:4])  # (Index 1 to 3)
print(colors[:2])   # (First two)
print(colors[-1])   # (Last item)
```

3. **List Comprehensions (The "Pro" Way)**


This is a concise way to create a new list based on an existing one.


```python
numbers = [1, 2, 3, 4, 5]

# Square every number in the list
squares = [n**2 for n in numbers]
print(squares) # Output: [1, 4, 9, 16, 25]

# Create a list of even numbers only
evens = [n for n in numbers if n % 2 == 0]
print(evens)
```

Tuple : A tuple in Python is a built-in data type used to store a collection of items in a single variable. It is an ordered, immutable (unchangeable), and indexed sequence.

Tuple Iteam are

1. ordered iteam as list with small breaket

2. Immutable: Once created, you cannot add, remove, or change items.

3. Duplicate is allowed here

4. Heterogeneous: A single tuple can hold different data types, such as strings, integers, and even lists.



**Basic Operations:**

Creation: Define items inside parentheses () separated by commas.

  1.Single-item Tuple: Requires a trailing comma, e.g., my_tuple = ("apple",).


Accessing Items: Use square brackets [] with an index starting at 0.

  1. Negative Indexing: -1 refers to the last item.


Slicing: Access a range of items using tuple[start:stop:step].


Packing & Unpacking: Assigning values to a tuple is "packing," while extracting them into variables is "unpacking

```python

fruits = ("apple", "banana", "cherry", "banana")

print(fruits[0])
print(fruits.count("banana"))
```

2. **Unpacking Tuples**


You can "unpack" the elements of a tuple directly into variables. This is a very common Pythonic pattern.

```python
point = (10, 20, 30)
x, y, z = point

print(f"x: {x}, y: {y}, z: {z}")
```

3.** Returning Multiple Values from a Function**


Functions often return tuples when they need to provide more than one result.

```python
def user():
    return "Alice", 25, "Engineer"

name, age, job = user()
print(name)
print(age)
print(job)
```

4. **Using Tuples as Dictionary Keys**
**bold text**


Because tuples are immutable, they can be used as keys in a dictionary, whereas lists cannot.


```python
locations_Capital = {
    ("India"): "New Delhi",
    (34.0522, -118.2437): "Los Angeles"
}

print(locations_Capital[("India")])
print(locations_Capital[(34.0522, -118.2437)])
```

5. **Swapping Variables**



Tuples allow you to swap the values of two variables in a single line without a temporary third variable.


```python
a = 5
b = 10

a, b = b, a
print(f"a: {a}, b: {b}")
```

**SET :**
a set is a built-in data structure used to store a collection of unique elements in a single variable. It is particularly useful for tasks where you need to eliminate duplicates or perform mathematical set operations.



**Key Characteristics**

**Unordered**: Elements do not have a defined order, so they may appear differently each time you print the set.


**Unique Elements:** Duplicate values are automatically removed. If you add the same item twice, it will only appear once.


**Unindexed:** Because they are unordered, you cannot access items by an index (e.g., my_set[0] will raise an error).


**Mutable:** You can add or remove items after the set is created, but the individual elements must be immutable (e.g., strings, numbers, or tuples)



**Basic Usage and Syntax :**

Sets are written with curly brackets {} or created using the set() constructor.


```python
# Creating a set with duplicates
my_set = {"apple", "banana", "cherry", "apple"}
print(my_set)  # Output: {'banana', 'apple', 'cherry'} (order may vary)

# Creating an empty set (MUST use set(), as {} creates an empty dictionary)
empty_set = set()

```

the most common set operations in Python, using both shorthand operators and built-in methods.


1. Basic Set Setup


```python
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}
```

2. **Union (All items from both)**

Combines elements from both sets, removing duplicates.


```python
# Operator: |
union_res = set_a | set_b
# Method: .union()
union_res = set_a.union(set_b)

print(union_res)
```

3.** Intersection (Only common items)**


Finds only the elements that exist in both sets.

```python
# Operator: &
intersect_res = set_a & set_b
# Method: .intersection()
intersect_res = set_a.intersection(set_b)

print(intersect_res)
```

4. **Difference (Unique to one side)**


Returns items in the first set that are not in the second.

```python
# Operator: -
diff_res = set_a - set_b
# Method: .difference()
diff_res = set_a.difference(set_b)

print(diff_res) # Output: {1, 2, 3}
```

5. **Symmetric Difference (Items NOT in both)**


Returns items that are in either set, but not in their intersection.

```python
# Operator: ^
sym_diff_res = set_a ^ set_b
# Method: .symmetric_difference()
sym_diff_res = set_a.symmetric_difference(set_b)

print(sym_diff_res)
```

6. **Membership and Modification**

```python
# Check if item exists (very fast in sets)
if 3 in set_a:
    print("3 is present")

# Adding and Removing
set_a.add(10)        # Adds 10
set_a.remove(1)      # Removes 1 (raises error if missing)
set_a.discard(99)
```
