3.Logical Operator : Logical operators are used to combine conditional statements:

**and** 	Returns True if both statements are true	x < 5 and  x < 10	.

**or**	Returns True if one of the statements is true	x < 5 or x < 4

**not**	Reverse the result, returns False if the result is true	not(x < 5 and x < 10)

```python
x = 5

print(x > 3 and x < 10)
```

```python
is_logged_in = False

# Reverses False to True
if not is_logged_in:
    print("Please log in to continue.")
```

```python
age = 25
has_ticket = True
is_banned = False

# 1. 'and' operator: Both conditions must be True
if age >= 18 and has_ticket:
    print("Access granted: You are an adult with a ticket.")

# 2. 'or' operator: At least one condition must be True
day = "Saturday"
if day == "Saturday" or day == "Sunday":
    print("It's the weekend!")

# 3. 'not' operator: Reverses the boolean value
if not is_banned:
    print("You are allowed to enter the venue.")

# 4. Combining operators (Precedence: not > and > or)
# This checks if someone is eligible for a discount:
# Must be (a student OR a senior) AND NOT banned.
is_student = True
is_senior = False

if (is_student or is_senior) and not is_banned:
    print("Discount applied!")
```

4. Bitwise Operator : In Python, bitwise operators perform operations on integers by treating them as sequences of binary bits (0s and 1s). They are primarily used in low-level programming, such as cryptography, performance optimization, and hardware interactions.

Operator 	Name	Description	Example (a=10, b=4)


**&	AND**	Sets bit to 1 if both bits are 1.	10 & 4 → 0


**|	OR**	Sets bit to 1 if at least one bit is 1.	10 | 4 → 14


**^	XOR	**Sets bit to 1 if bits are different.	10 ^ 4 → 14


**~	NOT**	Inverts all bits (flips 0 to 1 and vice versa).	~10 → -11


**<<**	Left Shift	Shifts bits left, filling with 0s on the right.	10 << 1 → 20


**>>**	Right Shift	Shifts bits right, discarding rightmost bits.	10 >> 1 →

```python
# Values in binary:
# a (10) = 1010
# b (4)  = 0100

a = 10
b = 4

# 1. Bitwise AND: Results in 0000 (0)
print(f"a & b: {a & b}")

# 2. Bitwise OR: Results in 1110 (14)
print(f"a | b: {a | b}")

# 3. Bitwise XOR: Results in 1110 (14)
print(f"a ^ b: {a ^ b}")

# 4. Bitwise NOT: Inverts bits (formula: -(x + 1))
# ~10 results in -11
print(f"~a: {~a}")

# 5. Bitwise Left Shift: 1010 becomes 10100 (20)
# Effectively multiplies by 2^n
print(f"a << 1: {a << 1}")

# 6. Bitwise Right Shift: 1010 becomes 0101 (5)
# Effectively floor divides by 2^n
print(f"a >> 1: {a >> 1}")
```

**5. Comparision Operator **
Comparison operators (also called relational operators) are used to compare two values. They evaluate the relationship between operands and return a Boolean value:

**either True or False. **

```python
x = 15
y = 20
name1 = "Ajit"
name2 = "Sonu"

# 1. Basic Equality
print(f"Is x equal to y? {x == y}")       # False

# 2. Inequality
print(f"Is x not equal to y? {x != y}")   # True

# 3. Size Comparisons
print(f"Is x > 10? {x > 10}")             # True
print(f"Is y <= 20? {y <= 20}")           # True

# 4. String Comparison (based on alphabetical order/Unicode)
print(f"Does 'Ajit' come before 'Sonu'? {name1 < name2}") # True

# 5. Chained Comparisons (a unique Python feature)
# Checks if 10 is less than x AND x is less than 20
is_between = 10 < x < 20
print(f"Is 10 < x < 20? {is_between}")    # True
```

6. IDENTITY OPERATOR :

dentity operators are used to check if two variables point to the exact same object in memory.
While comparison operators (==) check if values are equal, identity operators (is) check if the memory address (ID) is the same.


Operator 	Description	Example

**is**	Returns True if both variables point to the same object.	x is y


**is not**	Returns True if variables point to different objects.	x is not y


```python
# Create two identical lists
list_a = [1, 2, 3]
list_b = [1, 2, 3]
list_c = list_a  # c points to the same object as a

# 1. 'is' vs '=='
# They have the same values, so == is True
print(f"list_a == list_b: {list_a == list_b}")  # True

# They are different objects in memory, so 'is' is False
print(f"list_a is list_b: {list_a is list_b}")  # False

# 2. Assigning the same object
# list_c was set to list_a, so they share the same ID
print(f"list_a is list_c: {list_a is list_c}")  # True

# 3. 'is not' operator
print(f"list_a is not list_b: {list_a is not list_b}") # True

# 4. Checking Memory Addresses
print(f"ID of a: {id(list_a)}")
print(f"ID of b: {id(list_b)}")
```

7. MEMBERSHIP OPERATOR :
Membership operators are used to test whether a value or variable is found within a sequence, such as a string, list, tuple, set, or dictionary.

Operator	Description	Example
**in**	Returns True if the value is present in the sequence.	'a' in 'apple'


**not in**	Returns True if the value is not present in the sequence.	'z' not in 'apple'



```python
# 1. Using 'in' with a List
fruits = ["apple", "banana", "cherry"]
print(f"Is 'apple' in fruits? {'apple' in fruits}")     # True

# 2. Using 'not in' with a String
message = "Hello World"
print(f"Is 'z' not in message? {'z' not in message}")   # True

# 3. Membership in Dictionaries
# Note: This checks KEYS by default, not values.
user_data = {"name": "Alice", "age": 25}

print(f"Is 'name' in user_data? {'name' in user_data}") # True
print(f"Is 'Alice' in user_data? {'Alice' in user_data}") # False (Alice is a value)

# 4. In a conditional statement
guest_list = ["Bob", "Charlie", "David"]
current_user = "Eve"

if current_user not in guest_list:
    print(f"Access denied: {current_user} is not on the list.")
```
