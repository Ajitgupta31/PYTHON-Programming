## OOPs Concepts for Interview Notes

### 1. Class and Object

**Class:** A blueprint or a template for creating objects. It defines a set of attributes (data) and methods (functions) that the objects created from the class will have.

**Object:** An instance of a class. It is a real-world entity that has a state (attributes) and behavior (methods).



**Constructor (**__init__): A special method used to initialize an object's state.


**Self:** A reference to the current instance of the class, used to access variables belonging to that class

```python
class Dog:
    def __init__(self, name, age):
        self.name = name  # Instance attribute
        self.age = age

    def bark(self):       # Method
        return f"{self.name} says Woof!"

# Creating an object
my_dog = Dog("Buddy", 3)
print(my_dog.bark())
```

```python
class Dog:
    # Class attribute
    species = "Canis familiaris"

    # Initializer / Instance attributes
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # Instance method
    def description(self):
        return f"{self.name} is {self.age} years old"

    # Another instance method
    def speak(self, sound):
        return f"{self.name} says {sound}"

# Create objects (instances) of the Dog class
my_dog = Dog("Buddy", 5)
your_dog = Dog("Lucy", 2)

# Access attributes and call methods
print(f"My dog's name is {my_dog.name}.")
print(f"My dog is a {my_dog.species}.")
print(my_dog.description())
print(my_dog.speak("Woof!"))

print(f"\nYour dog's name is {your_dog.name}.")
print(your_dog.description())
print(your_dog.speak("Arf!"))
```

#### Another Class and Object Example: Car

```python
class Car:
    # Class attribute
    wheels = 4

    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.is_running = False

    def start_engine(self):
        if not self.is_running:
            self.is_running = True
            return f"The {self.make} {self.model}'s engine started."
        else:
            return f"The {self.make} {self.model}'s engine is already running."

    def stop_engine(self):
        if self.is_running:
            self.is_running = False
            return f"The {self.make} {self.model}'s engine stopped."
        else:
            return f"The {self.make} {self.model}'s engine is already off."

    def get_description(self):
        return f"This is a {self.year} {self.make} {self.model} with {self.wheels} wheels."

# Create car objects
car1 = Car("Toyota", "Camry", 2020)
car2 = Car("Honda", "Civic", 2022)

# Interact with car1
print(car1.get_description())
print(car1.start_engine())
print(car1.start_engine()) # Try starting again
print(f"Car1 running status: {car1.is_running}")

# Interact with car2
print(f"\n{car2.get_description()}")
print(car2.stop_engine()) # Try stopping before starting
print(car2.start_engine())
print(car2.stop_engine())
print(f"Car2 running status: {car2.is_running}")
```

### 2. Encapsulation

Encapsulation is the bundling of data (attributes) and methods (functions) that operate on the data into a single unit, or class. It also restricts direct access to some of an object's components, which means preventing unintended interference and misuse of data. This is often achieved using access modifiers (like private, protected, public, though Python uses conventions).



**Why Use Encapsulation?**

Encapsulation provides several benefits:


*   Data Protection: Prevents accidental modification of data

*   Validation: You can validate data before setting it

*   Flexibility: Internal implementation can change without affecting external code

*   Control: You have full control over how data is accessed and modified





```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.__age = age # Private property

p1 = Person("Email", 25)
print(p1.name)
#print(p1.__age) this will cause an error
```

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.__owner = owner  # 'private' attribute (by convention)
        self.__balance = balance # 'private' attribute (by convention)

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}. New balance: ${self.__balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.__balance}")
        else:
            print("Invalid withdrawal amount or insufficient funds.")

    def get_balance(self):
        return self.__balance

    def get_owner(self):
        return self.__owner

# Create an account
account = BankAccount("Alice", 100)

# Interact with the account using methods
account.deposit(50)
account.withdraw(30)
print(f"Current balance: ${account.get_balance()}")

# Attempting direct access (discouraged, but possible in Python)
# print(account.__balance) # This would raise an AttributeError
print(f"Owner: {account.get_owner()}")

# Python's 'name mangling' allows access via _ClassName__attributeName
# print(account._BankAccount__balance) # This works, but breaks encapsulation principle
```

**Name Mangling**

Name mangling is how Python implements private properties and methods.


When you use double underscores __, Python automatically renames it internally by adding _ClassName in front.


For example, __age becomes _Person__age.

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.__age = age

p1 = Person("Ajit", 30)

# This is how Python mangles the name:
print(p1._Person__age)
```

### 3. Inheritance

Inheritance is a mechanism that allows a new class (subclass/derived class) to inherit properties and behaviors (attributes and methods) from an existing class (superclass/base class). This promotes code reusability and establishes an 'is-a' relationship between classes.

**Create a Parent Class**

Any class can be a parent class, so the syntax is the same as creating any other class:

```python
class Person:
  def __init__(self, fname, lname):
    self.firstname = fname
    self.lastname = lname

  def printname(self):
    print(self.firstname, self.lastname)

#Use the Person class to create an object, and then execute the printname method:

x = Person("Ajit", "Gupta")
x.printname()
```

**Create a Child Class**

To create a class that inherits the functionality from another class, send the parent class as a parameter when creating the child class:

**Example**
Create a class named Student, which will inherit the properties and methods from the Person class:

class Student(Person):
  pass

```python
class Student(Person):
  pass  #Note: Use the pass keyword when you do not want to add any other properties or methods to the class.

x = Student("Sonu", "Kumar")
x.printname()
```

**Add the __init__() Function**

So far we have created a child class that inherits the properties and methods from its parent.


We want to add the __init__() function to the child class (instead of the pass keyword).


**Note: **The __init__() function is called automatically every time the class is being used to create a new object.

```python
class Animal:
   def __init__(self,name):
       self.name= name

   def speak(self):
       print(self.name)

# Create the Dog class (inherits from Animal)
class Dog(Animal):
   pass

# Create an object
d1= Dog("Rex")

# Call the speak method
d1.speak()
```

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        return f"{self.name} is eating."

    def sleep(self):
        return f"{self.name} is sleeping."

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name) # Call the constructor of the parent class
        self.breed = breed

    def bark(self):
        return f"{self.name} (a {self.breed}) barks loudly!"

class Cat(Animal):
    def __init__(self, name, color):
        super().__init__(name)
        self.color = color

    def meow(self):
        return f"{self.name} (a {self.color} cat) meows softly."

# Create objects
doggy = Dog("Max", "Golden Retriever")
kitty = Cat("Whiskers", "Tabby")

# Dog inherits from Animal
print(doggy.eat())
print(doggy.sleep())
print(doggy.bark())

# Cat inherits from Animal
print(kitty.eat())
print(kitty.sleep())
print(kitty.meow())
```

### 4. Polymorphism

Polymorphism means 'many forms'. In OOP, it allows objects of different classes to be treated as objects of a common base class. This is often achieved through method overriding (subclasses providing their own implementation of a method defined in the superclass) or method overloading (less common in Python, but means methods with the same name but different parameters).

**Method Overriding:** A subclass provides a specific implementation for a method that is already defined in its parent class.

**Function Polymorphism**

An example of a Python function that can be used on different objects is the len() function.

**String**
For strings len() returns the number of characters:

```python
x = "Hello World!"

print(len(x))
```

**Tuple**

For tuples len() returns the number of items in the tuple:

```python
fruit = ("apple", "banana", "cherry")

print(len(fruit))
```

**Dictionary**

For dictionaries len() returns the number of key/value pairs in the dictionary:

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

print(len(thisdict))
```

**Class Polymorphism**

Polymorphism is often used in Class methods, where we can have multiple classes with the same method name.

```python
class Cat:
   def sound(self):
    print ("meow")

# Create the Fox class

class Fox:
   def sound(self):
     print("Wa-pa-pa-pa-pa-pow!")

# Create objects and loop
c1= Cat()
f1= Fox()

for animal in (c1,f1):
   animal.sound()
```

```python
class Animal:
    def make_sound(self):
        return "Generic animal sound"

class Dog(Animal):
    def make_sound(self):
        return "Woof! Woof!"

class Cat(Animal):
    def make_sound(self):
        return "Meow!"

class Duck(Animal):
    def make_sound(self):
        return "Quack! Quack!"

# A function that takes an Animal object (or any object with a make_sound method)
def animal_noise(animal):
    print(animal.make_sound())

# Create objects of different classes
dog = Dog()
cat = Cat()
duck = Duck()

# Call the same function with different objects
animal_noise(dog)
animal_noise(cat)
animal_noise(duck)

# Even directly
print(dog.make_sound())
print(cat.make_sound())
```

### 5. Abstraction

Abstraction is the concept of hiding the complex implementation details and showing only the essential features of an object. It focuses on 'what' an object does rather than 'how' it does it. In Python, this is typically achieved using abstract base classes (ABCs) and abstract methods.

**Why use Abstraction?**


**Enforces a Contract:** It ensures that any programmer creating a subclass must implement the required methods (like calculate_area), or Python will raise an error.


**Reduces Complexity:** The user of the Circle class doesn't need to know how the math works internally; they just need to know that they can call .calculate_area().


**Prevents Instantiation: **You cannot create an object of an abstract class directly. For example, my_shape = Shape() would throw a TypeError.

```python
from abc import ABC, abstractmethod

class Shape(ABC): # Inherit from ABC to make it an abstract base class
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

    def describe(self):
        return "This is a generic shape."

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14159 * self.radius * self.radius

    def perimeter(self):
        return 2 * 3.14159 * self.radius

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

# Cannot instantiate an abstract class directly
# shape = Shape() # This would raise a TypeError

circle = Circle(5)
rectangle = Rectangle(4, 6)

print(f"Circle Area: {circle.area()}")
print(f"Circle Perimeter: {circle.perimeter()}")
print(circle.describe())

print(f"\nRectangle Area: {rectangle.area()}")
print(f"Rectangle Perimeter: {rectangle.perimeter()}")
print(rectangle.describe())
```
