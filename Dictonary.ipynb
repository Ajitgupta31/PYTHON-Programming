**Dictonary**

:a dictionary is a collection of key-value pairs. It works like a real-life dictionary or a phonebook: you look up a unique "key" (like a name) to find its associated "value" (like a phone number).




1. Dictonary is a mutable data type in python

2. A python dictonary is a collection of key and value pairs seperated by a colon(:)& enclosed in curly braces{}.

3. Keys must be unique in a dictonary, duplicate value is allowed.

```python
user = {
    "name": "Alice",
    "age": 25,
    "job": "Developer"
}

# Accessing a value by its key
print(user["name"])  # Output: Alice

# Using .get() (safer because it won't crash if the key is missing)
print(user.get("salary", "Not Provided"))
```

2. **Adding and Updating**


```python
user["age"] = 26

# Add a new key-value pair
user["city"] = "New York"
user["Phone"] = 1234567890

print(user)
```

3.** Removing Items**


```python
del user["job"]

# Remove and return the value
age = user.pop("age")

print(user)
```

4. **Looping through a Dictionary**


You can loop through keys, values, or both at once.


```python
car = {"brand": "Tesla", "model": "Model 3", "year": 2023}

# Loop through keys and values
for key, value in car.items():
    print(f"{key}: {value}")
```

5. **Dictionary Comprehension**


A fast way to create a dictionary from an iterable.


```python
squares = {x: x**2
           for x in range(1, 5)}
print(squares)
```

**Nesting Dictionaries inside a List**

This is exactly how real-world data (like JSON) usually looks. Imagine a list of users:

```python
users = [
    {"id": 1, "name": "Alice", "email": "alice@example.com"},
    {"id": 2, "name": "Bob", "email": "bob@example.com"},
    {"id": 3, "name": "Charlie", "email": "charlie@example.com"}
]
# Accessing the second user's name
print(users[1]["name"])
```
